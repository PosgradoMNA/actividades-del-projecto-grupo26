# **Parte 1: Fundamentos de bases de datos**

#### ***1. Fundamentos de bases de datos para ciencia de datos***

`RTA`: 
En ciencia de datos cuando nos referimos a bases de datos, estas son definidas como un conjunto de datos estructurados los cuales son almacenados en nubes, memorias u otro tipo de almacenamiento el cual pueda ser accesible. Las bases de datos se clasifican en dos principalmente, bases relaciones y bases no relacionales o también conocidas como NoSQL. Las bases relaciones o también conocidas como SQL (Structured Query Language) son aquellas que son almacenadas en tablas y pueden ser consultadas de manera sencilla con cualquier relación específica a esta y son clasificadas en dos, almacenamiento de tablas relacionales y OLAP analítico. Ejemplos de estas son tablas bancarias con transacciones, productos de clientes almacenadas en PostgreSQL. Las bases NoSQL se definen en función de la información por categorías del tipo clave-valor. Estas claves deben ser únicas donde se puedan realacionado a cada categoría. Esto se debe porque no solo se almacenan tablas, aquí se pueden almacenar documentos, información de redes sociales, imágenes siendo diferente a las tablas de las SQL. Para NoSQL existen cuatro tipos de bases: Bases de datos documentos, almacenamienito clave-valor, bases de datos orientadas a columnas y de bases de datos de gráficos. Ejemplos de SQL; ORACLE, MySQL, SQLSever, ejemplos de NoSQL; mongoDB, cassandra, Apache HBASE.

#### ***2. Fundamentos de almacenes de datos (Data Warehourse) para ciencia de datos***

`RTA`:
Data warehouse es básicamente una base de datos gigante que toma diferentes fuentes (base de datosc como la ejemplificada en la pregunta 1) y las integra para transformarlas en un solo esquema que permitan por medio de consultas o `queries` obtener información de las diferentes fuentes simultaneamente y a partir de esto generar todos los análisis y descripciones que sean pertinentes. Algo que es importante considerar es que una base de datos ORACLE puede ser considerado un data warehouse, ya que esta puede estar particionada en función del negocio. Este tipo de arquitecturas son muy comunes en empresas donde lo que se busca es poder centralizar mucha fuentes en un solo lugar para que las consultas sean lo más sencillas posible y poder establercer reglas que permitan hacer crecer el negocio.

## Selección y limpieza de datos

```python
path = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
df = pd.read_csv(path)
```

### ***Nombres de columnas***
Debido a que las columnas no poseen nombres originales, se procede a revisar la información de cada uno de los atributos y nombrando cada una de las columnas con su respectivo nombre

```python
col = ['ID', 'Amount of the given credit', 'Gender', 'Education', 'Marital status', 'Age', 'the repayment status in September', 'the repayment status in August', 
       'the repayment status in July', 'the repayment status in June', 'the repayment status in May', 'the repayment status in April',
       'Amount of bill statement in September, 2005', 'Amount of bill statement in August, 2005', 'Amount of bill statement in July, 2005', 
       'Amount of bill statement in June, 2005', 'Amount of bill statement in May, 2005', 'Amount of bill statement in April, 2005',
       'Amount paid in September, 2005', 'Amount paid in August, 2005', 'Amount paid in July, 2005', 'Amount paid in June, 2005', 'Amount paid in May, 2005',
       'Amount paid in April, 2005', 'Class']
df.columns = col
```

```python
df.head()
```

### ***Verficación de datos nulos***
Se verifica la cantidad de valores nulos que puede contener el conjuntos de datos

```python
df.info()
```

Se percibe que la variable `target` posee tres datos que no tiene valor. Imputar ese tipo de datos es una tarea difícil debido a que no se conoce el comportamiento de que pueda tener cada variable para definir este valor único en cada una de las filas. Así que se comienza con la verificación de esta variable.

```python
row_target = pd.isnull(df['Class'])
df_target = df[row_target]
df_target.head()
```

Como se evidencia en el conjuntos de datos, para las variables `Amount of bill statement` y `Amount of paid` posee valores nulos los cuales pueden ser imputados pero como hacen parte de aquellos valores del `target` que no tiene valor, se procede a eliminarlos.

```python
df_v1 = df[~row_target].reset_index(drop=True)
df_v1.info()
```

#### ***Variables categóricas***

Teniendo en cuenta que a pesar que las variables `Gender`, `Marital_status`, `Education` y `Age`, contiene valores numéricos, estas pueden ser consideradas categóricas. A partir de esto, comenzaremos el análisis de `Gender`. Verificamos la cantidad de datos total que contiene el conjunto y buscaremos el valor que hace falta para esta variable

```python
df_v1.shape
```

```python
row_gender = pd.isnull(df_v1['Gender'])
df_gender = df_v1[row_gender]
df_gender
```

Se evidencia que casi todas las varibles poseen valores nulos, así que se procede a eliminar esta fila

```python
df_v2 = df_v1[~row_gender].reset_index(drop=True)
```

Procedemos a verficar las otras columnas categóricas mencionadas arriba `Matiral status` y `Education`

```python
row_cat = pd.isnull(df_v2['Marital status'])
df_cat = df_v2[row_cat]
df_cat
```

```python
df_v2 = df_v2[~row_cat].reset_index(drop=True)
row_cat2 = pd.isnull(df_v2['Age'])
df_cat2 = df_v2[row_cat2]
df_cat2
```

Verificamos que solo son tres observaciones con valores nulos. Existen muchas formas de decisión a la hora de imputar este tipo de valores. Acá, se realizará una de ellas y es usar alguna de las variables que se infiera tiene un comportamiento que pueda representar aquella que poseen valores nulos. La alternativa rápida es borrar esos registros ya que solo representarian el 0.01% de las muestras, pero se decide optar por esta estrategía para poder aplicar en otras variables. Se tomará como base la variable `Amount of the given credit` y se marcaran rangos que puedan representar los valores de la `Age`. Primero se realiza un histograma para evidencia la distribución de esta y luego se procede a encontrar estos rangos.

```python
fig, axs = plt.subplots(figsize=(12, 8))
df_v2['Amount of the given credit'].hist(ax=axs)
axs.set_title('Distribución del monto del crédito [Amount of the given credit]')
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 8))
df_v2['Age'].hist(ax=axs)
axs.set_title('Distribución de la edad [Age]')
plt.show()
```

Creación de los rangos con los datos no nulos

```python
datos_no_nulos = df_v2[~row_cat2].reset_index(drop=True)
datos_nonulos = datos_no_nulos[['Amount of the given credit', 'Age']]
agrupamiento_nonulos = datos_nonulos.groupby(pd.cut(datos_nonulos['Amount of the given credit'], np.arange(int(datos_nonulos['Amount of the given credit'].min())\
    ,int(datos_nonulos['Amount of the given credit'].max())+10000, int(datos_nonulos['Amount of the given credit'].min())))).median()
agrupamiento_nonulos.head()
```

Ahora lo que haremos es imputar esas medianas de edades en las tres observaciones de `Age` que poseen valores nulos, usando el siguiente loop

```python
df_cat2 = df_cat2.reset_index(drop=True)
datos_imp = agrupamiento_nonulos.reset_index(drop=True)
row = 0
while row < df_cat2.shape[0]:
    for i in range(datos_imp.shape[0]-1):
        
        if df_cat2.loc[row, 'Amount of the given credit'] > datos_imp.loc[i, 'Amount of the given credit'] \
            and df_cat2.loc[row, 'Amount of the given credit'] <= datos_imp.loc[i+1, 'Amount of the given credit']:
            df_cat2.loc[row, 'Age'] = datos_imp.loc[i, 'Age']
    
    row = row+1
```

```python
df_cat2
```

```python
data_final = pd.concat([df_cat2, datos_no_nulos], axis=0)
```
